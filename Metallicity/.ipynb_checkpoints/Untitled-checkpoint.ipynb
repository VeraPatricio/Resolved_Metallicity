# Maps in source plane transformed from image plane

```python
%matplotlib notebook
import matplotlib.pylab as plt
from astropy.io import fits
import numpy as np
```

```python
as1063=fits.getdata('SourcePlane_AS1063_met.fits')
as1063[np.where(as1063<8.7)] = np.nan
plt.figure()
plt.imshow(as1063,origin='lower',vmin=8.7,vmax=8.95)
plt.axis('off')
```

```python
a370_head = fits.getdata('SourcePlane_met_head.fits')
a370_1 = fits.getdata('SourcePlane_met_part1.fits')
a370_2 = fits.getdata('SourcePlane_met_part2.fits')
a370_3 = fits.getdata('SourcePlane_met_part3.fits')
a370_head[np.where(a370_head<8.6)] = np.nan
a370_1[np.where(a370_1<8.6)] = np.nan
a370_2[np.where(a370_2<8.6)] = np.nan
a370_3[np.where(a370_3<8.6)] = np.nan


fig, ax = plt.subplots(2,2,figsize=(6,6))
ax = ax.ravel()
[x.axis('off') for x in ax]
ax[0].imshow(a370_head,origin='lower',vmin=8.7,vmax=8.95)
ax[1].imshow(a370_1,origin='lower',vmin=8.7,vmax=8.95)
ax[2].imshow(a370_2,origin='lower',vmin=8.7,vmax=8.95)
ax[3].imshow(a370_3,origin='lower',vmin=8.7,vmax=8.95)

ax[0].set_title('Head')
ax[1].set_title('Part 1')
ax[2].set_title('Part 2')
ax[3].set_title('Part 3')
```
