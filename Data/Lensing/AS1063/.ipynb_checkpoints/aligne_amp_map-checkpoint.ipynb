```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))
%matplotlib notebook
import matplotlib.pylab as plt
from astropy.io import fits
from reproject import reproject_exact,reproject_interp
```

```python
white_light = fits.getdata('/Users/vera/Arcs/AS1063/Data/WhiteLight_AS1063.fits')
header = fits.getheader('/Users/vera/Arcs/AS1063/Data/WhiteLight_AS1063.fits',ext=1)

amp_map,_ = reproject_interp('amp_best_AS1063.fits',header)

fig, ax = plt.subplots(1,2,figsize=(14,4))
ax[0].imshow(amp_map,origin='lower',vmin=1, vmax=20)
ax[1].imshow(white_light,origin='lower',vmin=0, vmax=1000)

# write down aligned amplification map
fits.writeto('amp_aligned_AS1063.fits',data=amp_map,header=header)
```
