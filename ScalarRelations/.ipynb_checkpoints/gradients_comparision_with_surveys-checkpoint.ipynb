```python
carton = Table.read('Carton2018/J_MNRAS_478_4293_tablea1.dat.fits',format='fits')
                   
carton.show_in_notebook()
```

```python
sns.set_style('darkgrid')
fig, ax = plt.subplots(1,1,figsize=(6,5))
fig.subplots_adjust(left=0.15,top=0.98,right=0.98)
ax.set_xlabel('Redshift')
ax.set_ylabel('$\Delta\,$Z (dex/ kpc)')
ax.set_xlim(0,1.2)

# to have a common hue:
z = np.concatenate((wuyts['z'].data,carton['z'].data))
mass = np.concatenate((wuyts['logM*'].data,carton['Mass'].data))
grad = np.concatenate((wuyts['dN2/dr'].data,carton['dlogZ'].data))
w = ['Wuyts+2016' for i in range(len(wuyts))]
c = ['Carton+2018' for i in range(len(carton))]
origin = np.concatenate((w,c))
  
mean_carton = np.mean(carton['dlogZ'][np.where((carton['z'] > 0.5) & (carton['z'] < 0.8))])
mean_wuyts = np.mean(wuyts['dN2/dr'][np.where((wuyts['z'] > 0.5) & (wuyts['z'] < 0.8) )])
std_carton = np.std(carton['dlogZ'][np.where((carton['z'] > 0.5) & (carton['z'] < 0.8))])
std_wuyts = np.std(wuyts['dN2/dr'][np.where((wuyts['z'] > 0.5) & (wuyts['z'] < 0.8) )])
mean_wuyts_1 = np.nanmean(wuyts['dN2/dr'][np.where((wuyts['z'] > 0.9) & (wuyts['z'] < 1.1) )])
std_wuyts_1 = np.nanstd(wuyts['dN2/dr'][np.where((wuyts['z'] > 0.9) & (wuyts['z'] < 1.1) )])

my_gal = ['AS1063-arc','A370-sys1','M1206-sys1']
my_z = [0.6,0.7,1.0]
my_mass = [10.94,10.40,10.90]
my_grad = [-0.042,-0.053,-0.037]
my_err = [0.002,0.004,0.060]

sns.scatterplot(x=z,y=grad,hue=mass,style=origin,markers=('o','s'),palette='Oranges',s=60)
ax.errorbar(my_z,my_grad,yerr=my_err,marker='*',markersize=15,alpha=0.9,color='0.1',elinewidth=0.7,linewidth=0,capsize=4)

# Add by hand
ax.annotate('M$_\odot$', xy=(0.2,0.235),zorder=10)
ax.annotate('M$_\odot$', xy=(0.2,0.206),zorder=10)
ax.annotate('M$_\odot$', xy=(0.2,0.180),zorder=10)
ax.annotate('M$_\odot$', xy=(0.22,0.152),zorder=10)
ax.annotate('M$_\odot$', xy=(0.22,0.125),zorder=10)

#ax.errorbar(0.66,mean_carton,yerr=std_carton,marker='s',markersize=8,alpha=0.9,color='0.3',elinewidth=0.7,linewidth=0,capsize=2)
#ax.errorbar(0.64,mean_wuyts,yerr=std_wuyts,marker='o',markersize=8,alpha=0.9,color='0.3',elinewidth=0.7,linewidth=0,capsize=2)
#ax.errorbar(1.01,mean_wuyts_1,yerr=std_wuyts_1,marker='o',markersize=8,alpha=0.9,color='0.3',elinewidth=0.7,linewidth=0,capsize=2)

print(mean_carton,'z=0.66')
print(mean_wuyts,'z=0.64')
print(mean_wuyts_1,'z=1.01')

plt.legend(loc='upper left')
fig.savefig('../Plots/lit_comparision.pdf')
```

Convert to the normalised units

```python
Re = [7.7,12.0,11.0]
print('normalised gradient',np.array(my_grad)*np.array(Re))

def error(Re,Re_err,grad,grad_err):
    return Re*grad_err + grad*Re_err

print('%0.3f'%error( 7.7,0.2,my_grad[0],my_err[0]))
print('%0.3f'%error(12.0,0.7,my_grad[1],my_err[1]))
print('%0.3f'%error(11.1,0.2,my_grad[2],my_err[2]))
```
